#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
fig, [ax1, ax2] = plt.subplots(1,2,figsize=(15,4) )

median_steps = []
month = []
for i in range(1, 13):
    temp = lo_que_me_interesa.loc[lo_que_me_interesa['Months_encoded'] == i]
    median_steps.append(temp['Steps'].median())
    month.append(i)

median_steps
month

ax1.plot(month, median_steps)
ax1.set_title('Median Steps By Month')   


days = lo_que_me_interesa['Days'].unique()
medianStepsDays = []
for day in days:
    temp = lo_que_me_interesa.loc[lo_que_me_interesa['Days'] == day]
    medianStepsDays.append(temp['Steps'].median())

ax2.plot(days, medianStepsDays)
ax2.set_title('Median Steps By Day') 

plt.show()
```
